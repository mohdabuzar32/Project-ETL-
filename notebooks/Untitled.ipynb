```python
import requests
import pandas as pd

schemes = {
    "SBI Bluechip": 119551,
    "ICICI Bluechip": 120503,
    "Nippon Large Cap": 118632,
    "Axis Bluechip": 119092,
    "Kotak Bluechip": 120841
}

all_data = []

for scheme_name, code in schemes.items():

    url = f"https://api.mfapi.in/mf/{code}"

    response = requests.get(url)
    data = response.json()

    df = pd.DataFrame(data["data"])

    df["scheme_name"] = scheme_name
    df["amfi_code"] = code

    all_data.append(df)

combined_df = pd.concat(all_data, ignore_index=True)

print(combined_df.head())

combined_df.to_csv("../data/raw/all_5_schemes_nav.csv", index=False)

print("All 5 schemes NAV saved successfully!")
```

```python
import pandas as pd

fund_master = pd.read_csv("../data/raw/01_fund_master.csv")
nav_history = pd.read_csv("../data/raw/02_nav_history.csv")

fund_codes = set(fund_master["amfi_code"])
nav_codes = set(nav_history["amfi_code"])

missing_codes = fund_codes - nav_codes

print("Missing Codes:", missing_codes)
print("Number of Missing Codes:", len(missing_codes))
```
